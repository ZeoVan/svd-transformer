### Tokkenizer 

```python
train.iloc[:,0] = train.iloc[:,0].apply(str)
```

```python
tokenizer = keras.preprocessing.text.Tokenizer(char_level=False,lower=False)
tokenizer.fit_on_texts(list(train.iloc[:,0]))
print('Number of tokens: ',len(tokenizer.word_counts))
```
